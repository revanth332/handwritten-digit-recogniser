```python
from PIL import Image,ImageChops
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = Image.open("my_image.jpg","r")
image = image.resize((28,28))
px_data = list(image.getdata())
px_data_latest = []
for item in px_data:
    px_data_latest.append(sum(item)/255)
image.show(px_data_latest)
```

```python
px_data_latest = np.array(px_data_latest)
plt.imshow(px_data_latest.reshape(28,28))
plt.show()
```
